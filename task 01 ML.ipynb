```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
#Load the Dataset
df = pd.read_csv("C:\\Users\\hp\\OneDrive\\Desktop\\machine learning\\archive (1)\\credit_risk_dataset.csv")

df.head()
```

```python
#Understand the Data
df.info()
df.isnull().sum()
```

```python
#Handle Missing Values
df['person_emp_length'].fillna(df['person_emp_length'].median(), inplace=True)
df['loan_int_rate'].fillna(df['loan_int_rate'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
#data preprocessing
#Separate Features (X) and Target (y)
x = df.drop('loan_status', axis=1)
y = df['loan_status']
```

```python
x.shape
```

```python
df['loan_status'].unique()
```

```python
#Encode Categorical Features
x = pd.get_dummies(x, drop_first=True)
x.head()
```

```python
#Train–Test Split
X_train, X_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
#Feature Scaling (ONLY FEATURES)
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

MODELING SECTION

Logistic Regression

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_scaled, y_train)

lr_pred = lr.predict(X_test_scaled)
lr_prob = lr.predict_proba(X_test_scaled)[:, 1]
lr_prob
lr.fit(X_train_scaled, y_train)
```

Decision Tree

```python
dt = DecisionTreeClassifier(
    max_depth=6,
    min_samples_split=10,
    random_state=42
)

dt.fit(X_train, y_train)

dt_pred = dt.predict(X_test)
dt_prob = dt.predict_proba(X_test)[:, 1]
dt.fit(X_train, y_train)
```

Random Forest

```python
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)
rf_prob = rf.predict_proba(X_test)[:, 1]
```

Model Evaluation Function

```python
def evaluate_model(name, y_test, y_pred, y_prob):
    print(f"\n{name}")
    print("Accuracy :", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred))
    print("Recall   :", recall_score(y_test, y_pred))
    print("F1 Score :", f1_score(y_test, y_pred))
    print("ROC-AUC  :", roc_auc_score(y_test, y_prob))
```

```python
evaluate_model("Logistic Regression", y_test, lr_pred, lr_prob)
evaluate_model("Decision Tree", y_test, dt_pred, dt_prob)
evaluate_model("Random Forest", y_test, rf_pred, rf_prob)
```

```python
results = pd.DataFrame({
    "Model": ["Logistic Regression", "Decision Tree", "Random Forest"],
    "Accuracy": [
        accuracy_score(y_test, lr_pred),
        accuracy_score(y_test, dt_pred),
        accuracy_score(y_test, rf_pred)
    ],
    "ROC-AUC": [
        roc_auc_score(y_test, lr_prob),
        roc_auc_score(y_test, dt_prob),
        roc_auc_score(y_test, rf_prob)
    ]
})

results
```

```python
#combine all methods to get a goos result 
from sklearn.ensemble import VotingClassifier
voting_clf = VotingClassifier(
    estimators=[
        ('lr', lr),
        ('dt', dt),
        ('rf', rf)
    ],
    voting='soft'  # Use 'hard' for majority voting
)
voting_clf.fit(X_train_scaled, y_train)  # Note: Use scaled features for Logistic Regression
voting_pred = voting_clf.predict(X_test_scaled)
voting_prob = voting_clf.predict_proba(X_test_scaled)[:, 1]
evaluate_model("Voting Classifier", y_test, voting_pred, voting_prob)
```
